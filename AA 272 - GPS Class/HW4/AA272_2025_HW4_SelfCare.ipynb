# Self-care

We are approaching the end of the autumn quarter (and this is your last homework for this class!)
For many of us, end-of-the-quarter deadlines can be especially stressful.
But, we also have the upcoming winter break to look forward to!

For this self-care practice, feel free to choose one of the following prompts to respond to:

- What is one thing, big or small, that you have recently done to take care of yourself during the end-of-the-quarter rush?

- After a particularly demanding period of time, when we ask a lot more of ourselves, it is vital to give ourselves extra time to rest and fully recover.
    In this respect, how do you plan to use the upcoming winter break to take extra good care of yourself?
    You all worked very hard this autumn quarter; you deserve it!

Describe your thoughts in a few sentences.
Alternatively, feel free to be creative in your expression, including but not limited to a photo, video, or sketch.
