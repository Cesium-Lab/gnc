```python
import sys
sys.path.append("..")

import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go

from csim.simulation import Simulator
from csim.entities import Spacecraft
from csim.world import R_EARTH, W_EARTH
from csim.constants import FT_TO_M
from csim.visualization import plot_orbit
```

```python
dt = 1
t0 = 0
n_steps = 3000

r = R_EARTH
state0 = np.array([
    0, 0, R_EARTH + 100e3,
    7000,0,800, 
    1,0,0,0, 
    0,0,0])

sc = Spacecraft(100, None)
sim = Simulator(state0, t0, dt, n_steps, sc)

sim.simulate()
```

```python
plot_orbit(sim.X)
```

```python

# plot orbits 
fig = go.Figure()

# Plot trajectory
fig.add_scatter3d(
    x=sim.X[:,0],
    y=sim.X[:,1],
    z=sim.X[:,2],
    mode="markers",
    marker=dict(
        size=1,
        color="blue"
    ),
    name="Trajectory"
)

# Final Position
fig.add_scatter3d(
    x=[sim.X[-1,0]],
    y=[sim.X[-1,1]],
    z=[sim.X[-1,2]],
    mode="markers",
    marker=dict(
        size=10,
        color="green"
    ),
    name="Final Position"
)


# Initial Position
fig.add_scatter3d(
    x=[sim.X[0,0]],
    y=[sim.X[0,1]],
    z=[sim.X[0,2]],
    mode="markers",
    marker=dict(
        size=10,
        color="red",
    ),
    name="Initial Position"
)

# Earth!
# X, Y, Z = makesphere(0, 0, 0, R_EARTH, 100)
# fig.add_surface(x=X, y=Y, z=Z, colorscale=[[0, "blue"], [1,"blue"]], showlegend=False, showscale=False, name="Earth")

# Temp Earth

# x = [[100, 100],
#      [0,   0]]

# y = [[ 1, -1],
#      [ 1, -1]]

# z = [[R_EARTH, R_EARTH],
#      [R_EARTH, R_EARTH]]

# fig.add_surface(x=x, y=y, z=z, colorscale=[[0, "orange"], [1,"orange"]],showlegend=False, showscale=False, name="Earth")

fig.update_layout(width=1000, height=800, 
                title_font=dict(size=24, family='Garamond'), title_text="Rocket Trajectory", title_x=0.5, title_y=0.9,
                # scene = dict(
                    # xaxis = dict(nticks=4, range=[-100,100]),
                    # yaxis = dict(nticks=4, range=[-50,100]),
                    # zaxis = dict(nticks=4, range=[R_EARTH-10,R_EARTH+10]))
                )
fig.show()
```
